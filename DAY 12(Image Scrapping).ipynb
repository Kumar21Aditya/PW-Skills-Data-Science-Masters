```python
import os
import requests
from bs4 import BeautifulSoup
import urllib

# URL of the website you want to scrape
url = "https://in.images.search.yahoo.com/search/images;_ylt=AwrKCwXqcVhlPncGXNW7HAx.;_ylu=Y29sbwNzZzMEcG9zAzEEdnRpZAMEc2VjA3BpdnM-?p=virat+kohli&fr2=piv-web&type=E210IN885G0&fr=mcafee"

# Create a directory to store the images
image_directory = "PW Skills Python/images"
os.makedirs(image_directory, exist_ok=True)

# Send an HTTP request to the URL
response = requests.get(url)

# Parse the HTML content of the page
soup = BeautifulSoup(response.text, "html.parser")

# Find image tags and extract URLs
image_tags = soup.find_all("img")
image_urls = [img["src"] for img in image_tags if "src" in img.attrs]

# Download images and save them to the specified directory
for i, img_url in enumerate(image_urls):
    # Create a unique filename for each image
    image_filename = os.path.join(image_directory, f"image_{i}.jpg")

    # Download and save the image
    urllib.request.urlretrieve(img_url, image_filename)

    print(f"Image {i + 1} saved to: {image_filename}")
```

```python
import os
import requests
from bs4 import BeautifulSoup
import urllib

# Disable SSL certificate verification (not recommended for production)
requests.packages.urllib3.disable_warnings(requests.packages.urllib3.exceptions.InsecureRequestWarning)

# URL of the website you want to scrape
url = "https://in.images.search.yahoo.com/search/images;_ylt=AwrKCwXqcVhlPncGXNW7HAx.;_ylu=Y29sbwNzZzMEcG9zAzEEdnRpZAMEc2VjA3BpdnM-?p=virat+kohli&fr2=piv-web&type=E210IN885G0&fr=mcafee"

# Create a directory to store the images
image_directory = "new_images"
os.makedirs(image_directory, exist_ok=True)

# Send an HTTP request to the URL, ignoring SSL verification
response = requests.get(url, verify=False)

# Parse the HTML content of the page
soup = BeautifulSoup(response.text, "html.parser")

# Find image tags and extract URLs
image_tags = soup.find_all("img")
image_urls = [img["src"] for img in image_tags if "src" in img.attrs]

# Download images and save them to the specified directory
for i, img_url in enumerate(image_urls):
    # Create a unique filename for each image
    image_filename = os.path.join(image_directory, f"image_{i}.jpg")

    # Download and save the image
    urllib.request.urlretrieve(img_url, image_filename)

    print(f"Image {i + 1} saved to: {image_filename}")
```
