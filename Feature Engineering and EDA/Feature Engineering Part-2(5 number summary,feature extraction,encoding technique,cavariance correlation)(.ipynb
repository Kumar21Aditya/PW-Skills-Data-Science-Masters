## Feature Scaling

```python
import seaborn as sns
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
import numpy as np
```

```python
mean=np.mean(df.total_bill)
std=np.std(df.total_bill)
print("mean is",mean)
print("standard deviation is",std)
```

```python
standardized_data=[]
for i in list(df.total_bill):
    val=(i-mean)/std
    standardized_data.append(val)
print(standardized_data)    
```

```python
sns.histplot(df.total_bill)
```

```python
sns.histplot(standardized_data)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(df[['total_bill']])
```

```python
scaler.transform(df[['total_bill']])
```

```python
scaler.fit_transform(df[['total_bill']])
```

```python
import pandas as pd
ss=pd.DataFrame(scaler.fit_transform(df[['total_bill','tip']]),columns=["bill_std","tips_std"])
ss
```

```python
import pandas as pd
df=pd.DataFrame({'city':['new york','london','paris','tokyo','new york','paris'],
                'price':[200,150,300,250,180,320]})
df
```
